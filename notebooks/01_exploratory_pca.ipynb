# 01 - Análise Exploratória com PCA

Este notebook contém a análise exploratória inicial dos dados espectrais de FTIR
para o projeto de autenticação de manteiga.

## 1. Importar bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

## 2. Carregar dados (exemplo com dados sintéticos)

**TODO:** Substituir pelos dados reais quando disponíveis.

```python
np.random.seed(42)
n_samples = 100
n_features = 100

X_pura = np.random.normal(0.5, 0.1, (n_samples//2, n_features))
X_adulterada = np.random.normal(0.3, 0.15, (n_samples//2, n_features))

X = np.vstack([X_pura, X_adulterada])
y = np.hstack([np.zeros(n_samples//2), np.ones(n_samples//2)])

df = pd.DataFrame(X)
df['classe'] = y
df['classe'] = df['classe'].map({0: 'Pura', 1: 'Adulterada'})

print(f"Shape dos dados: {X.shape}")
print(f"Distribuição das classes:\n{df['classe'].value_counts()}")
```

## 3. Visualizar os espectros (simulados)

```python
plt.figure(figsize=(12, 6))
for i in range(5):
    plt.plot(X[i], alpha=0.7, label=f'Pura {i+1}' if i < 3 else f'Adulterada {i-2}')
plt.xlabel('Comprimento de onda (simulado)')
plt.ylabel('Intensidade')
plt.title('Exemplo de espectros simulados')
plt.legend()
plt.show()
```

## 4. Padronizar os dados

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## 5. Aplicar PCA

```python
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

explained_variance = pca.explained_variance_ratio_

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.bar(range(1, len(explained_variance)+1), explained_variance)
plt.xlabel('Componente Principal')
plt.ylabel('Variância explicada')
plt.title('Variância explicada por componente')

plt.subplot(1, 2, 2)
plt.plot(range(1, len(explained_variance)+1), np.cumsum(explained_variance), marker='o')
plt.xlabel('Número de componentes')
plt.ylabel('Variância acumulada')
plt.axhline(y=0.95, color='r', linestyle='--', label='95%')
plt.legend()
plt.title('Variância acumulada')
plt.tight_layout()
plt.show()
```

## 6. Gráfico de scores (PC1 vs PC2)

```python
plt.figure(figsize=(8, 6))
colors = {'Pura': 'blue', 'Adulterada': 'red'}
for classe, color in colors.items():
    mask = df['classe'] == classe
    plt.scatter(X_pca[mask, 0], X_pca[mask, 1],
                c=color, label=classe, alpha=0.7, edgecolors='k')

plt.xlabel(f'PC1 ({explained_variance[0]:.2%} da variância)')
plt.ylabel(f'PC2 ({explained_variance[1]:.2%} da variância)')
plt.title('PCA - Manteiga Pura vs Adulterada')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.3)
plt.show()
```

## 7. Análise dos loadings (pesos das variáveis)

```python
loadings = pca.components_[0]

plt.figure(figsize=(12, 4))
plt.plot(loadings)
plt.xlabel('Variável (comprimento de onda simulado)')
plt.ylabel('Peso (loading)')
plt.title('Loadings da PC1')
plt.axhline(y=0, color='r', linestyle='--', alpha=0.5)
plt.grid(True, linestyle='--', alpha=0.3)
plt.show()

print("\n✅ Análise exploratória concluída!")
```
